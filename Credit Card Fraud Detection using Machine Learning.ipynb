# Credit_Card_Fraud_Detection using ML

# Data Summary:
This dataset consists of credit card transactions characterized by 31 features and a class label. The features capture various elements of each transaction, while the class label indicates whether the transaction is fraudulent (class 1) or legitimate (class 0).

The first feature, "Time," measures the number of seconds that have passed since the first transaction recorded in the dataset. Following this, there are 28 features labeled V1 to V28, which are anonymized variables generated through a principal component analysis (PCA) transformation of the original data. These variables reflect different aspects of the transactions, such as their amount, location, and type.

The penultimate feature, "Amount," indicates the transaction amount in USD, and the final feature is the "Class" label, which also marks the transaction as fraudulent (class 1) or not (class 0).

In summary, this dataset is designed for training machine learning models to identify fraudulent transactions in real-time. The features provide the necessary data for the model to learn patterns that can help flag potential fraud in future transactions.

The first line of code generates a new dataframe named "legit_transaction" by extracting rows from the original "credit_card" dataframe where the "Class" label equals 0. This effectively filters out fraudulent transactions (Class == 1) and retains only the legitimate ones (Class == 0).

The second line creates another dataframe called "fraud_transaction" which includes only the rows where the "Class" label is 1. This removes all legitimate transactions, focusing solely on those identified as fraudulent.

By dividing the data into these two separate dataframes, it becomes easier to examine and compare the traits of legitimate versus fraudulent transactions. This separation can be beneficial for uncovering patterns or features that are more prevalent in fraudulent cases, aiding in the development of effective fraud detection models.

```python
legit_transaction = credit_card[credit_card.Class==0]
fraud_transaction = credit_card[credit_card['Class']==1]
```

```python
from sklearn.preprocessing import RobustScaler
new_df_cc = credit_card.copy()
new_df_cc['Amount'] = RobustScaler().fit_transform(new_df_cc['Amount'].to_numpy().reshape(-1, 1))
time = new_df_cc['Time']
new_df_cc['Time'] = (time - time.min()) / (time.max() - time.min())
new_df_cc
```

```python
new_df_cc = new_df_cc.sample(frac=1, random_state=1)
new_df_cc
```

```python
X = new_df_cc.drop('Class', axis=1)
y = new_df_cc['Class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1
)
```

```python
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression()
logistic_model.fit(x_train, y_train)
logistic_model.score(x_train, y_train)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, logistic_model.predict(x_test), target_names=['Not Fraud', 'Fraud']))
```

Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions

Number of Fraudulent Transactions --> 492

A random sample of 492 observations from the legit dataset been taken. This is done to balance the number of observations in the legit and fraud datasets, which is necessary for training a machine learning model to predict fraud. Since the original dataset has a large number of legitimate transactions and a small number of fraudulent transactions, the model may be biased towards predicting that all transactions are legitimate. By creating a balanced dataset with an equal number of legitimate and fraudulent transactions, the model can be trained to better recognize the patterns that differentiate fraudulent transactions from legitimate ones

```python
X = new_df.drop(columns='Class', axis=1)
Y = new_df['Class']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

# Model Training

Logistic Regression

```python
model=LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy score on Test Data : ', test_data_accuracy)
```

## Model using tensorflow
As the model provided a good accuracy score, so Neural network using tensorflow will be implemented. 


```python
import tensorflow
```

```python



from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Dense, BatchNormalization
from tensorflow.keras.callbacks import ModelCheckpoint

# Create the neural network model
credit_nn = Sequential()
credit_nn.add(InputLayer((x_train.shape[1],)))
credit_nn.add(Dense(2, 'relu'))
credit_nn.add(BatchNormalization())
credit_nn.add(Dense(1, 'sigmoid'))

# Use the correct filepath for ModelCheckpoint
checkpoint = ModelCheckpoint('credit_nn.keras', save_best_only=True)

# Compile the model
credit_nn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
credit_nn.summary()
```

```python
credit_nn.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=5, callbacks=checkpoint)
```

```python
def neural_net_predictions(model, x):
  return (model.predict(x).flatten() > 0.5).astype(int)
neural_net_predictions(credit_nn, x_test)
```

```python
print(classification_report(y_test, neural_net_predictions(credit_nn, x_test), target_names=['Not Fraud', 'Fraud']))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(max_depth=2, n_jobs=-1)
rf.fit(x_train, y_train)
print(classification_report(y_test, rf.predict(x_test), target_names=['Not Fraud', 'Fraud']))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc = GradientBoostingClassifier(n_estimators=50, learning_rate=1.0, max_depth=1, random_state=0)
gbc.fit(x_train, y_train)
print(classification_report(y_test, gbc.predict(x_test), target_names=['Not Fraud', 'Fraud']))
```

```python
from sklearn.svm import LinearSVC
svc = LinearSVC(class_weight='balanced')
svc.fit(x_train, y_train)
print(classification_report(y_test, svc.predict(x_test), target_names=['Not Fraud', 'Fraud']))
```
